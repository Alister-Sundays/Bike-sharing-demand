## Step 3: Train a model using AutoGluon’s Tabular Prediction

Requirements:
* We are prediting `count`, so it is the label we are setting.
* Ignore `casual` and `registered` columns as they are also not present in the test dataset. 
* Use the `root_mean_squared_error` as the metric to use for evaluation.
* Set a time limit of 10 minutes (600 seconds).
* Use the preset `best_quality` to focus on creating the best model.

```python
# Define the features to be ignored during training
ignored_features = ["casual", "registered"]

# Pre-process the training data by removing unwanted features
train = train.drop(columns=ignored_features)

predictor = TabularPredictor(label = "count").fit(
    train_data=train,time_limit=600,presets="best_quality",
)
```

## Step 4: Exploratory Data Analysis and Creating an additional feature
* Any additional feature will do, but a great suggestion would be to separate out the datetime into hour, day, or month parts.

```python
# Create a histogram of all features to show the distribution of each one relative to the data. This is part of the exploritory data analysis
train.hist(figsize=(12, 10))  # Adjust figsize for better visualization
plt.tight_layout()
plt.show()
```

```python
# create a new feature
train["Hour"] = train["datetime"].dt.hour
train['Day'] = train['datetime'].dt.day
train['Month'] = train['datetime'].dt.month
train['Day_of_Week'] = train['datetime'].dt.day_of_week
train['Year'] = train['datetime'].dt.year
train['Time'] = train['datetime'].dt.time

test["Hour"] = test["datetime"].dt.hour
test['Day'] = test['datetime'].dt.day
test['Month'] = test['datetime'].dt.month
test['Day_of_Week'] = test['datetime'].dt.day_of_week
test['Year'] = test['datetime'].dt.year
test['Time'] = test['datetime'].dt.time
```

#### Time of Day Segmentation

```python
#Extract morning, afternoon, evening, and night, to capture different time-of-day patterns.

def get_time_of_day(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour < 24:
        return 'Evening'
    else:
        return 'Night'
```

```python
train['time_segment'] = train['Time'].apply(lambda x: get_time_of_day(x.hour))
test['time_segment'] = test['Time'].apply(lambda x: get_time_of_day(x.hour))
```

#### Day/Night Indicator

```python
#Create a binary feature to indicate whether it's daytime or nighttime based on the hour of the day
def get_day_or_night(hour):
    if 6 <= hour < 18:
        return 'Day'
    else:
        return 'Night'
```

```python
train['day_or_night'] = train['Hour'].apply(lambda x: get_day_or_night(x))
test['day_or_night'] = test['Hour'].apply(lambda x: get_day_or_night(x))
```

#### Cyclic Encodings

```python
def encode_hour(hour):
    #convert hour to radians
    hour_in_radians = hour / 24.0 * 2 * np.pi
    #Encode hour using sine and cosine transformations
    sin_hour = np.sin(hour_in_radians)
    cos_hour = np.cos(hour_in_radians)
    return sin_hour, cos_hour
```

```python
train['sin_hour'], train['cos_hour'] = zip(*train['Hour'].apply(encode_hour))
train['sin_hour'], train['cos_hour'] = zip(*train['Hour'].apply(encode_hour))

test['sin_hour'], test['cos_hour'] = zip(*test['Hour'].apply(encode_hour))
test['sin_hour'], test['cos_hour'] = zip(*test['Hour'].apply(encode_hour))

#We use the zip() function along with the apply() function to apply the encode_hour() function to each row of the 'hour' column, and then unpack the resulting tuples into separate columns
# 'sin_hour' and 'cos_hour' in the DataFrame.
```

```python
train.drop(columns=['Time'], inplace=True)
test.drop(columns=['Time'], inplace=True)
```

```python
train.head()
```

#### Make category types for these so models know they are not just numbers
* AutoGluon originally sees these as ints, but in reality they are int representations of a category.
* Setting the dtype to category will classify these as categories in AutoGluon.

```python
train["season"] = train["season"].astype("category")
train["weather"] = train["weather"].astype("category")
test["weather"] = test["weather"].astype("category")
test["season"] = test["season"].astype("category")
```

```python
# View are new feature
train.head()
```

#### Dropping unnecessary columns

```python
train.drop(columns=['datetime', 'registered','casual'],inplace=True)
test.drop(columns=['datetime'],inplace=True)
```

```python
test.head()
```

#### Create Dummy Variables for the season and weather columns


```python
season_type =  pd.get_dummies(train['season'],prefix='season')
season_type =  pd.get_dummies(test['season'],prefix='season')
season_type = season_type.astype(int)
season_type.head()
```

```python
weather_type = pd.get_dummies(train['weather'], prefix= 'weather')
weather_type = pd.get_dummies(test['weather'], prefix= 'weather')
weather_type = weather_type.astype(int)
weather_type.head()
```

```python
#Concatinating to the main dataframe
train = pd.concat([train, season_type, weather_type], axis=1)
test = pd.concat([test, season_type, weather_type], axis=1)

#Dropping the season and weather columns
train.drop(columns=['season', 'weather'], axis=1, inplace=True)
test.drop(columns=['season', 'weather'], axis=1, inplace=True)
```

```python
test.head()
```

```python
train['weekend_or_weekday'] = (train['Day_of_Week'] // 5 == 1).map({True: 'Weekend', False: 'Weekday'})

test['weekend_or_weekday'] = (test['Day_of_Week'] // 5 == 1).map({True: 'Weekend', False: 'Weekday'})
```

```python
train.drop(columns=['Day_of_Week'], axis=1, inplace=True)
test.drop(columns=['Day_of_Week'], axis=1, inplace=True)
```

```python
train.drop(columns=['workingday'], axis=1, inplace=True)
test.drop(columns=['workingday'], axis=1, inplace=True)
```

```python
#one-hot encode weekend_or_weekday column using get_dummies()
one_hot_encoded = pd.get_dummies(train['weekend_or_weekday'])
one_hot_encoded = pd.get_dummies(test['weekend_or_weekday'])
one_hot_encoded = one_hot_encoded.astype(int)
one_hot_encoded.head()

#Concatenate with the original DF
train = pd.concat([train, one_hot_encoded], axis=1)
test = pd.concat([test, one_hot_encoded], axis=1)

#drop the weekend_or_weekday column
train.drop(columns=['weekend_or_weekday'], inplace=True)
test.drop(columns=['weekend_or_weekday'], inplace=True)

train.head()
```

```python
train.head()
```

```python
#one-hot encode time_segment column using get_dummies()
one_hot_encoded1 = pd.get_dummies(train['day_or_night'])
one_hot_encoded1 = pd.get_dummies(test['day_or_night'])
one_hot_encoded1 = one_hot_encoded1.astype(int)
one_hot_encoded1.head()

#Concatenate with the original DF
train = pd.concat([train, one_hot_encoded1], axis=1)
test = pd.concat([test, one_hot_encoded1], axis=1)

#drop the weekend_or_weekday column
train.drop(columns=['day_or_night'], inplace=True)
test.drop(columns=['day_or_night'], inplace=True)

train.head()

```

## Step 6: Hyper parameter optimization
* There are many options for hyper parameter optimization.
* Options are to change the AutoGluon higher level parameters or the individual model hyperparameters.
* The hyperparameters of the models themselves that are in AutoGluon. Those need the `hyperparameter` and `hyperparameter_tune_kwargs` arguments.

```python
predictor_new_hpo = TabularPredictor(label = "count",).fit(
    train_data=train, time_limit=600, presets= "best_quality",
    hyperparameters={
        'RF': [
            {'n_estimators': 100, 'max_depth': 5},
            {'n_estimators': 200, 'max_depth': 8},
            {'n_estimators': 300, 'max_depth': 10},
        ]
    }
)
```
